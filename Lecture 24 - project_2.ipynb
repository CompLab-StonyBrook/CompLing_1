```python
import pandas as pd 
from collections import Counter
from pprint import pprint
from nltk.corpus import stopwords
import re


#the data is a csv file transferred from the json file from postman
data = pd.read_csv("/Users/ChongZhang/OneDrive/LB/most_frequent_words:phrases/json_to_csv/output.csv") 


def get_freq_words(arg): # a function to get the most common N words
    word_counter = Counter([string for i in arg for string in str(i).split()])
    # here you can change the number to any value
    popular_words = word_counter.most_common(100)
    return popular_words

def remove_stopwords(): # a function to remove functional words
    stop1 = set(stopwords.words('english'))
    # this set handles frequent yet uninformative words.
    # Perhaps this needs to be expanded constantly. 
    set2 = {'RT','-', "I'm", "You'll", 'via', 'here‚Ä¶', 'several', '&amp;', 'I', 'You', 
            'But', 'A', 'TO', 'IS', 'is', 'The', 'the', 'are'}
    stop = stop1 | set2
    return stop

def get_frequent_unigrams(): # a function to find the common words
    df_content = data["content"]
    df_content = df_content.as_matrix()
    return ([i for i in get_freq_words(df_content) if i[0] not in remove_stopwords()])

pprint("Some of the most common unigrams are: \n" +  str(get_frequent_unigrams()))

#a function to calculate the minimal edit distance between two strings
memo = {}
def get_LD(s1, s2):
    if len(s1) == 0:
        return len(s2)
    if len(s2) == 0:
        return len(s1)
    if s1[-1] == s2[-1]:
        cost = 0
    else:
        cost = 1
       
    i1 = (s1[:-1], s2)
    if not i1 in memo:
        memo[i1] = get_LD(*i1)
    i2 = (s1, s2[:-1])
    if not i2 in memo:
        memo[i2] = get_LD(*i2)
    i3 = (s1[:-1], s2[:-1])
    if not i3 in memo:
        memo[i3] = get_LD(*i3)
    d = min([memo[i1]+1, memo[i2]+1, memo[i3]+cost])
    
    return d


def get_aggregated_value(lst):
    # the list to be processed

    if len(lst) == 0:
        return []
    else:
        # start list of equivalence classes,
        # containing only equiv class for first element of lst;
        # each equivalence class is of the form [[word1, word2, ...], count]
        e_class = [[lst[0][0]], lst[0][1]]
        equiv = [e_class]
        # start iterating over rest of lst
        if len(lst) > 1:
            for (word, count) in lst[1:]:
                # check if we can find a matching equivalence class
                for e_class in equiv:
                    words = e_class[0]
                    element = words[0]
                    if get_LD(word, element) <= 2:
                        # add word to equivalence class
                        words.append(word)
                        # and increment counter
                        e_class[1] += count
                        break
                else:
                    # we haven't found any matching equivalence class
                    # so we add a new one
                    equiv.append([[word], count])
    return equiv


def cleanup_equiv_unigram(equiv):
    return [(re.sub("\W", "", words[0].upper()), count)
            for [words, count] in equiv]


print("The cleaned up unigram frequencies for word cloud are:", cleanup_equiv_unigram(get_aggregated_value(get_frequent_unigrams())))
```

```python
from collections import Counter
from pprint import pprint

def get_freq_ngrams(arg, n): # a function to find the most common 50 ngrams
    stack = [string for i in arg for string in str(i).split()]
    ngrams = [tuple(stack[pos:pos+n]) for pos in range(len(stack)-(n-1))]    
    return Counter(ngrams).most_common(50)

def get_freq_bigrams(arg): # a function to get bigrams
    return get_freq_ngrams(arg, 2)

def get_frequent_bigrams(): # a function to find the common bigrams in HEADLINE_ALERT_TEXT
    df_content = data["content"]
    df_content = df_content.as_matrix()
    return ([i for i in get_freq_bigrams(df_content) if i not in remove_stopwords()])
pprint("Some of the most common bigrams in 'get_frequent_bigrams' are in this list: " + str(get_frequent_bigrams()))
```

```python
import pandas as pd 
import numpy as np 
from pandas import Series,DataFrame

from gensim.models import Phrases
from gensim.models import Word2Vec

data = pd.read_csv("/Users/ChongZhang/OneDrive/LB/most_frequent_words:phrases/json_to_csv/output.csv") # read in data
data_description = data["content"] # specify the column HEADLINE_ALERT_TEXT
data_description = data_description.as_matrix() # convert to matrix
    
sentence = [str(doc).split(" ") for doc in data_description] # get the sentence stream
bigram = Phrases(sentence, min_count=1, threshold=2) # build phrases, i.e. bigrams in this case

#sent = data_cleaned_news[0].split(" ")
#print(bigram[sent])

#print(list(bigram[sentence])[:5])


bigram_model = Word2Vec(bigram[sentences], size=100) # load the model 
bigram_model_counter = Counter() # calculate the most common 100 phrases
for key in bigram_model.wv.vocab.keys():
    if key not in stopwords.words("english"):
        if len(key.split("_")) > 1:
            bigram_model_counter[key] += bigram_model.wv.vocab[key].count

for key, counts in bigram_model_counter.most_common(100): # this takes a bit of time to run
    print(key, counts)
```
